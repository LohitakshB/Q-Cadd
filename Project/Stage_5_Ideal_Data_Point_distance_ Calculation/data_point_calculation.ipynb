```python
import pandas as pd
import numpy as np
```

```python
filepath = r"\Qe-CaDD\Project\Data\Stage_4_QSVM_data\predicted_toxicity_of_ligands.csv"

df = pd.read_csv(filepath)

x_ideal = -12

y_ideal = 0

x_affinities = df["Affinity"].values

y_toxicity = df.iloc[:, 19:].values
```

```python

toxicity_scores_dist = []

dist_datapoints = []



for index, toxicity_scores in enumerate(y_toxicity):
    dist_datapoints = []
    for index, value in enumerate(toxicity_scores):
        x = x_affinities[index]
        y = value
        dist = ((x_ideal-x)**2 + (y_ideal - y)**2)**0.5

        dist_datapoints.append(float(dist))

    toxicity_scores_dist.append(dist_datapoints)
```

```python
print(toxicity_scores_dist)
```

```python
import statistics
avg_dist = []
for values in toxicity_scores_dist:

    avg = statistics.mean(values)
    avg_dist.append(avg)

print(len(avg_dist))
```

```python
df_dist = pd.DataFrame(avg_dist, columns=['Average Distance'])

df_combined = pd.concat([df, df_dist], axis=1)

df_combined.to_csv('C:/Qe-CaDD/Project/Data/Stage_5_datapoint_calculation_data/Final_toxicity_affinity_distance.csv', index=False)

df_combined.head()
```
