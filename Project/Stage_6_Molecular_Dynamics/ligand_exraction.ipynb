```python
import os
import pandas as pd

csv_path = rf'C:\Qe-CaDD\Project\Data\Stage_5_datapoint_calculation_data\Final_toxicity_affinity_distance.csv'
dict_data = rf'C:\Qe-CaDD\Project\Data\Stage_3_molecular_docking_data\Docking_data'

df = pd.read_csv(csv_path)


output_dir = rf'C:\Qe-CaDD\Project\Data\Stage_6_Molecular_Dynamics\ligands_top'
os.makedirs(output_dir, exist_ok=True)

for file in os.listdir(dict_data):
    if file in df["File"].values: 
        name_1,ext = file.split(".")
        name = f"{name_1}_extracted.pdbqt"
        path = os.path.join(output_dir, name)

        with open(os.path.join(dict_data, file), 'r') as f:
            lines = f.readlines()

        is_in_model_1 = False
        extracted_lines = []

        for line in lines:
            if line.startswith("ROOT"):
                is_in_model_1 = True
            if line.startswith("MODEL 2"):
                break
            if is_in_model_1:
                extracted_lines.append(line)




        with open(path, "w") as out_file:
            out_file.writelines(extracted_lines)
    else:
        continue

```

```python

import subprocess

for file in os.listdir("C:\Qe-CaDD\Project\Data\Stage_6_Molecular_Dynamics\ligands_top"):
    name , ext = file.split(".")
    full_file = rf"C:\Qe-CaDD\Project\Data\Stage_6_Molecular_Dynamics\ligands_top\{file}"
    path = rf'C:\Qe-CaDD\Project\Data\Stage_6_Molecular_Dynamics\ligands_sdf_top\{name}.sdf'


    result = subprocess.run(['obabel', full_file, '-O', path] , capture_output=True, text=True)


    if result.returncode == 0:
        print(f"Converted {file} to SDF format successfully.")
    else:    
        print(f"{result.stderr.strip()}")
```
