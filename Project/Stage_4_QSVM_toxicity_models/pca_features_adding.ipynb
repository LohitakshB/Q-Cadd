```python
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np
import os
from sklearn.decomposition import PCA
import pandas as pd
```

```python

def get_fingerprint(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        fingerprint = rdMolDescriptors.GetMorganFingerprintAsBitVect(mol, radius =2, nBits = 1024)
        return list(fingerprint)
    else:
        return None
```

```python
dict_data = r'C:/Qe-CaDD/Project/Stage_4_QSVM_toxicity_models/pathways'
```

```python
for files in os.listdir(dict_data):
    file_path = rf'C:/Qe-CaDD/Project/Stage_4_QSVM_toxicity_models/pathways/{files}'
    name, ext = files.split(".")
    df = pd.read_csv(file_path)
    data = []

    fingerprint = [get_fingerprint(smiles) for smiles in df['SMILES']]

    fingerprint_df = pd.DataFrame(fingerprint)

    pca = PCA(n_components= 16)

    reduced_fingerprints = pca.fit_transform(fingerprint_df)

    reduced_df = pd.DataFrame(reduced_fingerprints)



    df_tox = pd.read_csv(file_path)

    df_smiles = df_tox["SMILES"]

    df_output = df_tox[name]

    combined_df = pd.concat([df_smiles, reduced_df, df_output], axis=1)

    combined_df.to_csv(f'tox21_pca_{name}.csv', index=False)

    print(f'{combined_df} is saved')


```
