```python
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np
import os
from sklearn.decomposition import PCA
import pandas as pd
```

```python
def get_fingerprint(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        print(f"Invalid SMILES: {smiles}")
        return None
    else:
        fingerprint = rdMolDescriptors.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=2048)
    return list(fingerprint)
```

```python
file_path = rf"C:\Qe-CaDD\Project\Data\Stage_3_molecular_docking_data\docking_output_data.csv"

df = pd.read_csv(file_path)
data = []

fingerprint = [get_fingerprint(smiles) for smiles in df['SMILES']]

fingerprint_df = pd.DataFrame(fingerprint)

pca = PCA(n_components= 16)

reduced_fingerprints = pca.fit_transform(fingerprint_df)

reduced_df = pd.DataFrame(reduced_fingerprints)

combined_df = pd.concat([df,reduced_df],axis= 1)


print(combined_df)
```

```python
import os
import joblib

directory_path = rf"C:\Qe-CaDD\Project\Stage_4_QSVM_toxicity_models\model_svm"
for file in os.listdir(directory_path):
    name, ext = file.split(".")
    model = joblib.load(rf"C:\Qe-CaDD\Project\Stage_4_QSVM_toxicity_models\model_svm\{file}")
    predictions = model.predict(reduced_df)

    combined_df[name + "_predictions"] = predictions


combined_df.to_csv("C:\Qe-CaDD\Project\Data\Stage_4_QSVM_data\predicted_toxicity_of_ligands.csv" ,index= False)

print("Saved CSV to predicted_toxicity_of_ligands.csv")
    
```
